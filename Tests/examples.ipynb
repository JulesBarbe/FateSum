```python
# EXAMPLE SET 2
# SAVE TO CSV THEN EXPORT TO EXCEL

df = pd.DataFrame({"Task":[], "Dataset":[], "Model":[], "Text":[], "Summary":[], "Notes":[]})

task = []
dataset = []
model = []
text = []
summary = []
notes = []

# 40 examples with cnn_dm
with open('data/cnndm-bart-2159-s', 'rb') as f:
    c = sample(pickle.load(f), 10)
    for item in c:
        task.append("News")
        dataset.append("CNN/DM")
        model.append("BART-Large")
        text.append(item["text"])
        summary.append(item["summary_text"])
        notes.append("")
with open('data/cnndm-pegasus-2192-s', 'rb') as f:
    c = sample(pickle.load(f), 10)
    for item in c:
        task.append("News")
        dataset.append("CNN/DM")
        model.append("PEGASUS")
        text.append(item["text"])
        summary.append(item["summary_text"])
        notes.append("")

with open('data/cnndm-presummext-all', 'rb') as f:
    c = sample(pickle.load(f), 10)
    for item in c:
        task.append("News")
        dataset.append("CNN/DM")
        model.append("Extractive PRESUMM")
        text.append(item["text"])
        summary.append(item["summary_text"])
        notes.append("")

with open('data/cnndm-presummabs-all', 'rb') as f:
    c = sample(pickle.load(f), 10)
    for item in c:
        task.append("News")
        dataset.append("CNN/DM")
        model.append("Abstractive PRESUMM")
        text.append(item["text"])
        summary.append(item["summary_text"])
        notes.append("")


# 30 examples from xsum
with open('data/xsum-pegasus-2376-s', 'rb') as f:
    c = sample(pickle.load(f), 10)
    for item in c:
        task.append("News")
        dataset.append("XSUM")
        model.append("PEGASUS")
        text.append(item["text"])
        summary.append(item["summary_text"])
        notes.append("")

with open('data/xsum-presummext-all', 'rb') as f:
    c = sample(pickle.load(f), 10)
    for item in c:
        task.append("News")
        dataset.append("XSUM")
        model.append("Extractive PRESUMM")
        text.append(item["text"])
        summary.append(item["summary_text"])
        notes.append("")

with open('data/xsum-presummabs-all', 'rb') as f:
    c = sample(pickle.load(f), 10)
    for item in c:
        task.append("News")
        dataset.append("XSUM")
        model.append("Abstractive PRESUMM")
        text.append(item["text"])
        summary.append(item["summary_text"])
        notes.append("")


r = pd.DataFrame({"Task":task, "Dataset":dataset, "Model":model, "Text":text, "Summary":summary, "Notes":notes})
with pd.ExcelWriter("samples/summ_examples2.xlsx") as wr:
    r.to_excel(wr)


   

```
